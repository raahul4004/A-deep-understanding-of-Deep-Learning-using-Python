```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
```

Gradient descent in 1D

```python
#function (as a function)
def fx(x):
    return 3*x**2 - 3*x + 4

def deriv(x):
    return 6*x - 3
```

```python
#plt the function and its derivative

#define a range for x
x = np.linspace(-2,2,2001)
plt.plot(x,fx(x),x,deriv(x))
plt.xlim([x[0],x[-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()
```

```python
#random starting point 
np.random.seed(44)
localmin = np.random.choice(x,1)

#learning parameters
learning_rate = 0.01
training_epochs = 250

#run through training
for i in range(training_epochs):
    grad = deriv(localmin)
    localmin = localmin - grad*learning_rate
localmin
```

```python
#plotting the results
plt.plot(x,fx(x),x,deriv(x))
plt.plot(localmin, deriv(localmin), 'ro')
plt.plot(localmin, fx(localmin), 'ro')

plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x)','min'])
plt.title('Empirical local minimum: %s'%localmin[0])
plt.show()

```

```python
#random starting point 
#np.random.seed(44)
localmin = np.random.choice(x,1)

#learning parameters
learning_rate = 0.01
training_epochs = 150

#run through training
modelparams = np.zeros((training_epochs,2), dtype=float)

for i in range(training_epochs):
    grad = deriv(localmin)
    localmin = localmin - grad*learning_rate
    modelparams[i][0] = localmin
    modelparams[i][1] = grad

```

```python
#plt the gradient over iterations

fix, ax = plt.subplots(1,2,figsize = (12,4))

for i in range(2):
    ax[i].plot(modelparams[:,i],'o-')
    ax[i].set_xlabel('iteration')
    ax[i].set_title(f'Final estimated minimum: {localmin[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')
plt.show()
```

Coding Challenge

```python
#function 
def fx(x):
    return np.cos(2*np.pi*x) + x**2

def derivative(x):
    return -2*np.pi*np.sin(2*np.pi*x) + 2*x
```

```python
x = np.linspace(-2,2,2000)
local_min = np.array([0])

#learning
lr = 0.01
epochs = 100

for i in range(epochs):
    gradient = derivative(local_min)
    local_min = local_min - gradient*lr

local_min
```

```python
#plotting the results
plt.plot(x,fx(x),x,derivative(x))
plt.plot(local_min, derivative(local_min), 'ro')
plt.plot(local_min, fx(local_min), 'ro')

plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x)','min'])
plt.title('Empirical local minimum: %s'%local_min[0])
plt.show()
```
